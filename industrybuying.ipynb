```python
import requests
from bs4 import BeautifulSoup

for page in range(1,10):
    my_url = "https://www.industrybuying.com/brands/abb/motors-power-transmission-548/?page=1".format(page)
    res = requests.get(my_url).text
    soup1 = BeautifulSoup(res, "lxml")
    
    container = soup1.find_all("div",attrs={"class":"AH_ProductView col-lg-3 col-md-3 col-sm-6 col-xs-6 productThumbnails"})

    filename = "product2.csv"
    f = open(filename,"w")
    headers = "Product,Price,SKU,No\n"
    f.write(headers)

    for containers in container:
        title = containers.find("a",attrs={"class":"prFeatureName"})

        titles = title.text

        page = "https://www.industrybuying.com"
        title1 = title.get("href")
        title2 = page + title1

        title3 = requests.get(title2)
        soup2 = BeautifulSoup(title3.content,"html.parser")
        row = soup2.find_all("div",attrs= {"class":"filterRow"})

        for rows in row:
            tab1 = rows.find("div",attrs={"class":"featureNamePr"})
            tab2 = rows.find("div",attrs={"class":"featureValuePr"})

            tab1a = tab1.text.replace("\n","").replace(" ","")
            tab2a = tab2.text.replace("\n","").replace(" ","")

            #print("{}{}{}{}{}".format(titles,"|",tab1a,"|",tab2a))
            f.write(titles+"|"+tab1a+"|"+tab2a+"\n")
    f.close()
```
